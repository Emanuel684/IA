```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
dataset = pd.read_csv("Social_Network_Ads 1.csv")
```

```python
dataset.head()
```

```python
dataset = dataset.drop(["User ID"], axis=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset['Gender'] = dataset['Gender'].replace({'Male':1, 'Female':0}, regex=True)
```

```python
dataset.info()
```

```python
features = ['Gender','Age','EstimatedSalary']
```

```python
X = dataset[features]
y = dataset['Purchased']  #y = dataset.Purchased
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=3)
```

```python
clf.fit(X_train,y_train)
```

Dibujar el árbol de decisión

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True,feature_names=X.columns,fontsize=10)
plt.show()
```

Predicciones

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

Matriz de confusión

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
cm
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_pred)
```
