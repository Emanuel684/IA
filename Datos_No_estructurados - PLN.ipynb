```python
from nltk.tokenize import word_tokenize

token=tokenized_word=word_tokenize(text)
token
```

Reemplazar signos de puntuacion con vacio

```python
import string
token2=[]

for word in token:
    for letter in word:
      if letter in string.punctuation:
        word=word.replace(letter,"")
    token2.append(word)
token2
```

Quitar los vacios

```python
token3=[]

for word in token2:
  if word != "":
    token3.append(word)

len(token3)
```

Volver a todas minusculas

```python
token4=[]

for word in token3:
  word = word.lower()
  token4.append(word)

print(token4)
```


Quitar palabras con 1 o 2 carateres

```python
token5 = []
for word in token4:
  if len(word) >= 3:
    token5.append(word)

len(token5)
```

Stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
a=set(stopwords.words('english'))
stop_words = stopwords.words('english')
token6=[word for word in token5 if word not in a]
len(token6)
```

freqdist.
Drecuencia de las palabras

```python
from nltk.probability import FreqDist

fdist = FreqDist(token6)
fdist
```
